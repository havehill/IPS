# 고령인구 수 전처리

```python
old_data = pd.read_csv("/content/drive/MyDrive/2022_IPS/data/시군구별_고령인구수.csv", encoding="cp949")
old_data
```

```python
old_data.drop(['성별', '세대구성별'], axis=1, inplace=True)
old_data.drop(0, axis=0, inplace=True)
```

```python
mask = old_data['행정구역별(시군구)'].isin(['읍부', '면부', '동부'])
mask
```

```python
old_data = old_data[~mask]
old_data
```

```python
cido = yoyang_data['시도명'].unique()
print(cido)
print(len(cido))
```

```python
old_data.reset_index(drop=True, inplace=True)
old_data
```

```python
cidoInd = []
for i in range(len(old_data)):
  if old_data['행정구역별(시군구)'][i] in cido:
    cidoInd.append(i)

print(cidoInd)
print(len(cidoInd))
```

```python
old_data.drop(cidoInd, axis=0, inplace=True)
old_data
```

```python
old_data.rename(columns = {'행정구역별(시군구)' : '시군구명'}, inplace = True)
old_data
```

```python
old_grouped = old_data.groupby(['시도별', '시군구명'])
old_grouped.sum()
```

```python
old_total = pd.DataFrame(old_grouped.sum())
old_total
```

# 누적 확진자 수 전처리

```python
pat_data = pd.read_csv("/content/drive/MyDrive/2022_IPS/data/시군구별_누적확진자수.csv", encoding="cp949")
pat_data
```

```python
pat_data.loc[pat_data.지역.str.contains('고양시'), '지역'] = '고양시'
pat_data.loc[pat_data.지역.str.contains('성남시'), '지역'] = '성남시'
pat_data.loc[pat_data.지역.str.contains('수원시'), '지역'] = '수원시'
pat_data.loc[pat_data.지역.str.contains('안산시'), '지역'] = '안산시'
pat_data.loc[pat_data.지역.str.contains('안양시'), '지역'] = '안양시'
pat_data.loc[pat_data.지역.str.contains('용인시'), '지역'] = '용인시'
pat_data.loc[pat_data.지역.str.contains('청주시'), '지역'] = '청주시'
pat_data.loc[pat_data.지역.str.contains('천안시'), '지역'] = '천안시'
pat_data.loc[pat_data.지역.str.contains('포항시'), '지역'] = '포항시'
pat_data.loc[pat_data.지역.str.contains('전주시'), '지역'] = '전주시'
pat_data.loc[pat_data.지역.str.contains('창원시'), '지역'] = '창원시'
```

```python
pat_data
```

```python
pat_data.groupby(["시도명", "지역"])['확진자수'].sum()
```

```python
pat_total = pd.DataFrame(pat_data.groupby(["시도명", "지역"])['확진자수'].sum())
pat_total
```
